# Scalars, Vectors and Spaces

* `Scalars` are a single numerical value.
For example : 1,2,3 etc

* `Vectors` are ordered list of scalars.
For example: [1,2], [1,2,3], [-1,1] etc.

Let $v$ be a vector that represents the features of a room in a house say width of 20 and height of 20. Then we can represent the feature vector for that room as:-

$v=[20,10]$


```python
import matplotlib.pyplot as plt
```

```python
plt.plot(20, 10, 'bo')
plt.xlabel("Width")
plt.ylabel("Height")
plt.title("2D Vector Representation");
```
